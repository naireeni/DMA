```python
def fill_region (C):
    C.interpolate(method='linear',axis=1,inplace=True)
    C['% Signups']=(C['Signups']/C['Installs'])*100
    C['CAC']=C['Spend']/C['Signups']
    C.loc[:,('Total','Spend')]=C['Spend'].sum(axis=1)
    C.loc[:,('Total','Installs')]=C['Installs'].sum(axis=1)
    C.loc[:,('Total','Signups')]=C['Signups'].sum(axis=1)
    C.loc[:,('Total','% Signups')]=(C['Total']['Signups']/C['Total']['Installs'])*100
    C.drop(('Total','% Signed up'),axis=1,inplace=True)
    C.loc[:,('Total','CAC')]=C['Total']['Spend']/C['Total']['Signups']
    C.loc[:,('Total','CPI')]=C['Total']['Spend']/C['Total']['Installs']
    C.index=C.index.astype(str)
    C=C.fillna(0)
    
```

```python
fill_region(US)
fill_region(CA)
fill_region(UK)
fill_region(FR)
```
